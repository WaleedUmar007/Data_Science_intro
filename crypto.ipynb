```python
days= ["Monday","Tuesday","Wednesday","Thursday","Friday"]
KHR_temp= [25,22,21,26,22]
GJR_temp= [15,28,21,27,23]
plt.plot(days,KHR_temp,label='Temperature', color='r',marker='o')
plt.plot(days,GJR_temp,label='Temperature', color='b',marker='o',linestyle="--")
plt.legend(loc=4)
plt.xlabel("DAYS", fontsize=14)
plt.ylabel("TEMP", fontsize=14)

```

```python
days= ["Monday","Tuesday","Wednesday","Thursday","Friday"]
KHR_temp= [25,22,21,26,22]
GJR_temp= [15,28,21,27,23]
width=0.3
index = np.arange(len(days))
plt.bar(index,KHR_temp,width,label='Temperature', color='r',alpha=0.9,edgecolor='b',linewidth=1.5)
plt.bar(index+width,GJR_temp,width,label='GJRTemperature', color='b',edgecolor='r')
#since we have number of x-axis and want to show name of days we will do the following
plt.xticks(index+width,days, rotation=15)
plt.xlabel("DAYS", fontsize=14)
plt.ylabel("TEMP", fontsize=14)
plt.legend(loc=1)
```
